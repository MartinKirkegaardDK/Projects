## Map visualization

```python
data = pd.read_csv("../Data/T_T100D_SEGMENT_US_CARRIER_ONLY_2013_All.csv",delimiter=",")
latlong = pd.read_csv("../Data/uscities.csv",delimiter = ",")
latlong = latlong[["lat","lng","city","state_id"]]
```

```python
#data.columns
```

```python
li = []
for city, state in zip(latlong["city"],latlong["state_id"]):
    string = city + ", "+state
    li.append(string)
```

```python
latlong["ORIGIN_CITY_NAME"] = li
```

```python
#latlong
```

```python
latlong.drop(["city","state_id"],axis=1, inplace=True)
```

```python
x = data[["ORIGIN_CITY_NAME","DEST_CITY_NAME","PAYLOAD","DISTANCE","SEATS","PASSENGERS","DEPARTURES_PERFORMED"]]
```

```python
result = pd.merge(x,latlong, left_on = "ORIGIN_CITY_NAME",right_on="ORIGIN_CITY_NAME")
result["ORIGIN_COORDINATES"] = list(zip(result["lat"],result["lng"]))
result.drop(["lat","lng"],axis=1, inplace=True)
```

```python
result = pd.merge(result,latlong, left_on = "DEST_CITY_NAME",right_on="ORIGIN_CITY_NAME")
```

```python
result["DEST_COORDINATES"] = list(zip(result["lat"],result["lng"]))
result.drop(["lat","lng","ORIGIN_CITY_NAME_y"],axis=1, inplace=True)
result["ORIGIN_CITY_NAME"] = result["ORIGIN_CITY_NAME_x"]
result.drop(["ORIGIN_CITY_NAME_x"],axis=1, inplace=True)
```

```python
gdf = geopandas.GeoDataFrame(result)
```

```python
grouped = result.groupby(["ORIGIN_CITY_NAME","DEST_CITY_NAME"]).agg({"PAYLOAD":"sum","DISTANCE":"mean","SEATS":"sum","PASSENGERS":"sum","DEPARTURES_PERFORMED":"sum"})
```

```python
r = lambda: random.randint(0,255)
colors = set()
city_colors = dict()
for elm in grouped.iterrows():
    
    c = '#%02X%02X%02X' % (r(),r(),r())
    if elm[0][0] not in city_colors:
        while True:
            if c not in colors:
                city_colors[elm[0][0]] = c
                colors.add(c)
            break
```

```python
map1 = folium.Map()
```

```python
with open('../data/cb_2018_us_state_500k/cb_2018_us_state_500k.geojson') as f:
    gj = geojson.load(f)
```

```python
gdf_coord_name = gdf[['ORIGIN_COORDINATES','ORIGIN_CITY_NAME']]
```

```python
gdf_coord = gdf[['ORIGIN_COORDINATES','DEST_COORDINATES']].drop_duplicates()
gdf_coord =gdf_coord.sample(frac = 1)
```

```python
#gdf_coord
```

```python
for origin,name in zip(gdf_coord_name['ORIGIN_COORDINATES'],gdf_coord_name['ORIGIN_CITY_NAME']):
    folium.Marker(location= origin ,popup = name,icon=folium.Icon(color='lightgray', icon='plane',prefix = 'fa')).add_to(map1)
```

```python
#figure a way to add arrows
for points in zip(gdf_coord['ORIGIN_COORDINATES'][:100],gdf_coord['DEST_COORDINATES'][:100]):
    folium.PolyLine(points,color='black',symbol = 'arrow').add_to(map1)
```

```python

data = {

    -8:
        [
        0.547, 0.402, 0.591
        ],

    -4:
        [
        0.590, 0.462, 0.603
        ],
    0:
        [
        0.596, 0.486, 0.601
        ],

    4:
        [
        0.612, 0.465, 0.606
        ],

    8:
        [
        0.590, 0.458, 0.614
        ],

}

no_ablation = [
    0.607, 0.470, 0.609
]
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data preparation
datasets = ["Angry Tweets (Danish)", "ScaLA (Danish)", "Tweet Sentiment Extraction (English)"]
methods = data.keys()
scores = data

x = np.arange(len(datasets))  # Dataset indices
width = 0.15  # Width of bars

# Plotting with 'coolwarm' colormap and legend below
fig, ax = plt.subplots(figsize=(10, 6))
cmap = plt.cm.coolwarm
colors = [cmap(i / (len(scores) - 1)) for i in range(len(scores))]  # Generate colors

for i, (method, score) in enumerate(scores.items()):
    bars = ax.bar(x + i * width, score, width, label=method, color=colors[i], alpha=0.7)
    # Adding text on top of bars
    for bar in bars:
        ax.text(
            bar.get_x() + bar.get_width() / 2,  # Center of the bar
            bar.get_height(),  # Height of the bar
            f"{bar.get_height():.3f}",  # Value formatted to 3 decimal places
            ha='center', va='bottom', fontsize=9  # Alignment and font size
        )

# Customization
ax.set_xlabel("Datasets")
ax.set_ylabel("F1 Score")
ax.set_title("F1 Scores Across Different Methods and Datasets")
ax.set_xticks(x + width * (len(methods) - 1) / 2)
ax.set_xticklabels(datasets)

# Move the legend below the figure
ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.1), ncol=5, title='Fixed value for neurons')

fig.tight_layout()
fig.savefig('figures/downstream_tasks/bars.pdf', bbox_inches='tight')
plt.show()
```

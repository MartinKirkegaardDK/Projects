```python
import networkx as nx
import matplotlib.pyplot as plt
import mplleaflet
```

```python
G = nx.random_k_out_graph(15,2,alpha = 0.5)
my_pos = nx.spring_layout(G, seed = 100)
root = 2
edges = nx.bfs_edges(G, root)
nodes = [root] + [v for u, v in edges]
nodes
```

```python
nx.draw(G,pos = my_pos,with_labels = True,font_color = 'white')
```

```python
color_dict = dict()
for i in G.nodes():
    if i in nodes:
        color_dict[i] = 'red'
    else:
        color_dict[i] = 'green'
```

```python
colors = list(color_dict.values())
```

```python
nx.draw(G,pos = my_pos,node_color = colors,with_labels = True,font_color = 'white')
```
